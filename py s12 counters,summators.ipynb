```python
import random
import string
dir(string)
```

```python
even_count = 0 #counter
odd_count = 0 
summation = 0 #summator
even_sum = 0
odd_sum = 0
for i in range (0,5):
    num = random.randint(1,100)
    if (num % 2 == 0):
        even_count = even_count+1
        summation = summation + num
        even_sum = even_sum + num
    else:
        odd_count = odd_count+1
        summation = summation + num
        odd_sum = odd_sum + num

if even_count > odd_count:
    print("the even nums are more than odds")
else:
    print("the odd nums are more than evens")
print(f"the no of even no are :{even_count}")
print(f"the no of odd no are :{odd_count}")
print(f"the summation of the nums are :{summation}")
print(f"the even summation of the nums are :{even_sum}")
print(f"the odd summation of the nums are :{odd_sum}")
```
